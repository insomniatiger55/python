```python
from kafka import KafkaConsumer,TopicPartition
import json
```

```python
#json_deserilizer doesn't work
def json_deserilizer(data):
    return json.loads(data)
```

```python
topic='user_info'
consumer=KafkaConsumer(bootstrap_servers='localhost:9092',
                       group_id='my-group',auto_offset_reset='latest')
consumer.subscribe(topics=(topic))

#value_deserializer=json_deserilizer
```

```python
for message in consumer:
    info=json.loads(message.value)
    print ("Record inserted into topic : %s ,partition: %s, offset: %s " % (message.topic, message.partition,message.offset))
    print("Name: %s \nAddress: %s" 
          % (info['name'],info['address']))
```

# Reply from begining of specific partition

```python
TP=TopicPartition(topic,1)
consumer=KafkaConsumer(bootstrap_servers='localhost:9092',
                       auto_offset_reset='earliest')
#using assign have to pass in a list [] topicpartition
consumer.assign([TP])
consumer.seek_to_beginning()
```

```python
for message in consumer:
    info=json.loads(message.value)
    print ("Record inserted into topic : %s ,partition: %s, offset: %s " % (message.topic, message.partition,message.offset))
    print("Name: %s \nAddress: %s" 
          % (info['name'],info['address']))
```
